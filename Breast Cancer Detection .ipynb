### Breast Cancer Detection using Logistic Regression

#### Breast Cancer can be determined based on the type of tumor. There are 2 types of tumors-
#### 1. Benign Tumor: non-cancerous, slow growing and do not spread to other parts of the body.
#### 2. Malignant Tumor: cancerous, fast growing and spread to other parts of the body.

#### The type of tumor can be disgnosed based on the 'Fine Needle Procedure', the sample collected using this biopsy procedure is tested for Benign or Malignant Tumor based on some features like mean radius, mean texture,  mean perimeter, mean area, mean smoothness, mean compactness and so on that appear in the following dataset.  

```python
#importing required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset from sklearn

breast_cancer_data= sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_data)
```

```python
#loading the data into a pandas dataframe

df= pd.DataFrame(breast_cancer_data.data,columns=breast_cancer_data.feature_names)
```

```python
# printing the first five entries of the dataframe

df.head()
```

```python
# adding the target column as 'Diagnosis'

df['Diagnosis']= breast_cancer_data.target
```

```python
# checking if the target column is added

df.head()
```

```python
# checking the size of dataset

df.shape
```

```python
# getting some basic information about the dataset

df.info()
```

```python
# checking if data has some missing values

df.isnull().sum()
```

```python
# understanding some basic statistical measures about the data

df.describe()
```

```python
# checking distribution of target variable 'Diagnosos'
# here 1-->Benign and 0-->Malignant

df['Diagnosis'].value_counts()
```

```python
sns.countplot(x='Diagnosis',data=df)
```

```python
# grouping by the target variable and analyzing the ,ean values for both the cases

df.groupby('Diagnosis').mean()
```

```python
# separating the feature and target variables
# features are all the columns that contain information about the cells
# target variable is 'Diagnosis' which determines whether the sample is Malignant or Benign

X= df.drop(columns='Diagnosis',axis=1)
Y= df['Diagnosis']
```

```python
X.head()
```

```python
Y.head()
```

```python
# splitting the training and testing data

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
# checking the dimensions of the splitted datasets with the original dataset

print(X.shape,X_train.shape,X_test.shape)
```

```python
#loading an instance of Logistic Regression model in the variable

LR=LogisticRegression()
```

```python
# training Logistic Regression model using Training data

LR.fit(X_train,Y_train)
```

```python
# model evaluation
# accuracy on training data

X_train_pred=LR.predict(X_train)
training_accuracy=accuracy_score(Y_train,X_train_pred)
```

```python
print("Accuracy on Training Data is ",training_accuracy)
```

```python
# accuracy on testing data

X_test_pred=LR.predict(X_test)
test_accuracy=accuracy_score(Y_test,X_test_pred)
```

```python
print("Accuracy on Training Data is ",test_accuracy)
```
